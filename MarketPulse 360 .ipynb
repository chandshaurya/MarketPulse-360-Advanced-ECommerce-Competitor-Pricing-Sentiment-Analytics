```python
# outlier detection

# IQR

def detect_outliers_iqr(df, columns):
    outlier_summary = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_summary[col] = {
            "Q1": Q1,
            "Q3": Q3,
            "IQR": IQR,
            "Lower Bound": lower_bound,
            "Upper Bound": upper_bound,
            "Outlier Count": outliers.shape[0],
            "Outlier %": (outliers.shape[0] / df.shape[0]) * 100
        }
    return pd.DataFrame(outlier_summary).T


# column analysis
columns_to_check = {
    "Products": (prod, ["BasePrice"]),

    "Sale": (sale, ["Revenue", "OurPrice","CompetitorPrice"])
}

outlier_reports = {}

for name, (df, cols) in columns_to_check.items():
    report = detect_outliers_iqr(df, cols)
    outlier_reports[name] = report
    print(f"\n=== Outlier Analysis for {name} ===")
    print(report)

# Visualize all selected columns in subplots
for name, (df, cols) in columns_to_check.items():
    fig, axes = plt.subplots(nrows=1, ncols=len(cols), figsize=(6 * len(cols), 4))
    
    if len(cols) == 1:
        axes = [axes]  
    
    for ax, col in zip(axes, cols):
        sns.boxplot(x=df[col], ax=ax)
        ax.set_title(f"{name} - {col}", fontsize=12)
    
    plt.suptitle(f"Outlier Boxplots for {name}", fontsize=14, y=1.05)
    plt.tight_layout()
    plt.show()


```

```python
### CHURN Table


# Get the last transaction date per customer
last_txn = sale.groupby("CustomerID")['Date'].max().reset_index()
last_txn.rename(columns={'Date': 'last_txn_date'}, inplace=True)

# Merge with customers
cust_churn = pd.merge(cust, last_txn, on="CustomerID", how="left")

snapshot_date = sale['Date'].max()  # e.g., latest date in data

# Add churn column
cust_churn['days_since_last_txn'] = (snapshot_date - cust_churn['last_txn_date']).dt.days
cust_churn['churn_status'] = cust_churn['days_since_last_txn'].apply(
    lambda x: "Churned" if x > 90 else "Active"
)


Churn_table = cust_churn[["CustomerID", "Name", "last_txn_date", "days_since_last_txn", "churn_status","State","City","LoyaltyTier"]]
print(Churn_table.head())
```

```python
## Unique  vs Repeat Buyers

cust_purchase= sale.groupby('CustomerID')['ProductID'].count().reset_index()
cust_purchase.columns= ['CustomerID','PurchaseCount']

unique_buyers= (cust_purchase['PurchaseCount']==1).sum()

repeat_buyers= (cust_purchase['PurchaseCount']>1).sum()

total_cust = len(cust_purchase)

print("unique buyers: ",unique_buyers)
print("repeat buyers: ",repeat_buyers)

repeat_purchase_ratio = repeat_buyers/total_cust

print("Repeat Purchase Percentage: ", repeat_purchase_ratio *100)

plt.bar(['Unique buyers', 'Repeat buyers'],[unique_buyers,repeat_buyers],color=['skyblue','orange'])
plt.title("Unique  vs Repeat Buyers")
plt.ylabel("No. of Customers")

plt.show()
```

##  Pricing & Revenue Insights
- The business generated over **$30 billion in revenue** from more than **1 million units sold**.  
- The **average discount offered is ~15%**, with Social Media and Paid channels showing the deepest discounts, while Organic and Referral channels rely less on discounting.  
- **OurPrice** remains consistently lower than competitor prices, helping maintain a **competitive edge**. However, deep discounts also impact margins, highlighting the need for targeted discounting.  

##  Product & Category Insights
- **Electronics and Home Appliances** are the premier product categories, driving the majority of revenue.  
- The **top 10 most profitable products** are predominantly from the Electronics category, reinforcing its dominance.  
- The **overall return rate is exceptionally low at 0.06%**, reflecting strong product quality and high customer satisfaction. However, products with higher returns also show **lower-than-average ratings**.  
- Older SKUs continue to sell well, indicating that **product lifecycle longevity** is strong and not solely dependent on new launches.  

##  Regional Insights
- **Andhra Pradesh, Bihar, and Rajasthan** are the top three revenue-contributing states.  
- These regions represent strongholds for the business and opportunities for further campaign scaling.  


##  Sentiment & Review Insights
- The **average customer rating is 3.46 stars**, suggesting a generally positive experience with room for improvement.  
- Sentiment distribution shows the majority of reviews are **neutral to positive**, though negative reviews highlight issues with **delivery delays and product quality**.  
- There is a **weak correlation between high ratings and high sales volume**, suggesting that **brand and price are stronger purchase drivers** than ratings alone.  

##  Growth & Acquisition Insights
- Revenue shows a **consistent upward trend throughout the year**, peaking in the final quarter.  
- **Organic acquisition** is the most effective channel, bringing in the largest share of unique customers.  
- **Email marketing** attracts fewer customers, but they have a **higher average revenue per person**, making this channel high-value.  
- **New customer registrations peak in June and July**, suggesting seasonal demand patterns or successful mid-year campaigns.  

#  Executive Summary
- **Retention Gap**: While acquisition is strong, retention is weaker — loyalty initiatives are crucial.  
- **Discount Trade-off**: Blanket discounts drive sales but reduce margins — targeted, dynamic pricing is needed.  
- **Product Portfolio**: Electronics dominate revenue, bestsellers should be prioritized, while weak SKUs may be phased out.  
- **Customer Sentiment**: Ratings are generally positive, but operational issues like delivery and quality must be addressed.  
- **Competitive Edge**: OurPrice undercuts competitors, but long-term differentiation should focus on **trust, variety, and service quality** rather than only pricing.
